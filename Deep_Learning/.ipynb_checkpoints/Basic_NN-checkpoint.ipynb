### Defining the linear layer

```python
class linear :
    def __init__(self, in_neurons, out_neurons):
        self.in_neurons = in_neurons
        self.out_neurons = out_neurons
        self.weights = np.random.randn(out_neurons, in_neurons)*0.01 + 1
        self.bias = np.random.randn(out_neurons,1)
        
        self.dW = 0
        self.db = 0
        
        self.lastip = None
    
    def forward(self, input):
        self.lastip = input
        return np.dot(self.weights, input) + self.bias
    
    def backward(self, prevG, lr) : 
        ishape = self.weights.shape
        dW = np.dot(prevG, np.transpose(self.lastip))/self.lastip.shape[1]
        db = (np.sum(prevG, axis = 1, keepdims=True)/self.lastip.shape[1])
        self.weights = self.weights - lr * dW
        self.bias = self.bias - lr * db
        oshape = self.weights.shape
        if ishape != oshape:
            print("Changed Here")
        return np.dot(np.transpose(self.weights), prevG)
    
    def __call__(self, input):
        return self.forward(input)

net = linear(10, 20)
print(net(np.zeros(shape=(10,1))).shape)
x = net.backward(np.zeros(shape=(20, 1)), 0.1)
del(net)
```

### Activation functions
Let's define some activation functions. 

```python
class ActivationFunction :
    def __init__(self):
        return
    
    def forward(self, input):
        return input
    
    def __call__(self, input):
        return self.forward(input)

class ReLU(ActivationFunction):
    def __init__(self):
        self.lastinp = None
        return
    
    def forward(self, input):
        self.lastinp = input
        return np.maximum(0, input)
    
    def backward(self, prevG):
        self.lastinp[self.lastinp>0] = 1
        self.lastinp[self.lastinp<=0] = 0
        return np.multiply(prevG, self.lastinp)
    
class LeakyReLU(ActivationFunction):
    def __init__(self, slope=0.01):
        self.slope = slope
        self.lastinp = None
        return 
    
    def backward(self, prevG):
        self.lastinp[self.lastinp>0] = 1
        self.lastinp[self.lastinp<=0] = self.slope
        return prevG*self.lastinp
    
    
    
    def forward(self, input):
        self.lastinp = input
        return np.maximum(self.slope*input, input)
    
class Sigmoid(ActivationFunction):
    def __init__(self):
        self.lastinp = None
        return 
    
    def forward(self, input):
        self.lastinp = input
        self.lasty = 1/(1+np.exp(-input))
        return 1/(1+np.exp(-input))
    
    def backward(self, prevG):
        return np.multiply(np.multiply(self.lasty, (1-self.lasty)), prevG)
        

class Signum(ActivationFunction):
    def __init__(self):
        self.lastinp = None 
        return 
    
    def forward(self, input):
        self.lastinp = input 
        input[input>0] = 1
        input[input<=0] = 0
        return input
    
    def backward(self, prevG):
        self.lastinp[self.lastinp>0] = 1
        self.lastinp[self.lastinp<=0] = 0
        return np.multiply(self.lastinp, prevG)
```

#### Defining Loss Functions 
Implementing some standard loss functions. 

```python
class Loss():
    def __init__(self):
        self.loss = 0
        
    def findLoss(self, predicted, labels):
        assert predicted.shape == labels.shape, "Shape of Predicted labels and actual labels should be same"
        self.loss = 0 
        return self.loss
    
    def __call__(self, predicted, labels):
        return self.findLoss(predicted, labels)

loss = Loss()
loss(np.zeros(2), np.zeros(3))
```

### A sample neural network

```python
class Network1(Module):
    def __init__(self):
        super().__init__()
        self.l1 = linear(2, 3)
        self.sig1 = ReLU()
        self.l2 = linear(3, 1)
        self.sig2 = Sigmoid()
    
    def forward(self, input):
        if input.shape[1] == 2:
            input = input.T
        x = self.l1(input)
        x = self.sig1(x)
        x = self.l2(x)
        x = self.sig2(x)
        return x
    
    def predict(self, input):
        return self.forward(input)
    
    def backward(self, prevG):
        prevG = self.sig2.backward(prevG)
        prevG = self.l2.backward(prevG, 0.01)
        prevG = self.sig1.backward(prevG)
        prevG = self.l1.backward(prevG, 0.01)

net1 = Network1()
x = net1(np.random.randn(2,10))
x = net1(np.random.randn(2, 10))
net1.backward(np.random.randn(1,10))
print(x)

del(net1)
```

```python
 def load_extra_datasets():  
    N = 10000
    gaussian_quantiles = sklearn.datasets.make_gaussian_quantiles(mean=None, cov=0.7, n_samples=N, n_features=2, n_classes=2,  shuffle=True, random_state=None)
    return  gaussian_quantiles
gaussian_quantiles= load_extra_datasets()
X, Y = gaussian_quantiles
Y =  Y.reshape(Y.shape[0], 1)
# Visualize the data
plt.scatter(X[:, 0], X[:, 1], c=Y[:, 0], s=40, cmap=plt.cm.Spectral)
```

```python
X = X.reshape(10000,2)
net1 = Network1()
for epoch in range(1000):
    loss = 0 
    for i in range(1000):
        x = X[i*10:(i+1)*10, :].T
        op = net1(x)
        y = Y[i*10:(i+1)*10].T
        loss = -((1-y[0])*np.log(1-op)+ (y[0])*np.log(op))
        grad = (np.divide(1-y[0], (1-op[0])) - np.divide(y[0],op[0]))
        grad[np.isnan(grad) == True] = 1
        if math.isnan(np.sum(grad)) == False:
            net1.backward(grad)
        else :
            print("Not Updated")
    print("epoch : ", epoch, " loss : ", np.sum(loss))
```

```python
loss = 0
count = 0
plx, ply = [], []
plx1, ply1 = [], []
for i in range(9000, 10000):
    x = X[i].reshape(2,1)
    op = net1(x)
    op = op >= 0.5
    if op[0]== Y[i]:
        count+=1
    elif Y[i] == [1]:
        plx.append(x[0])
        ply.append(x[1])
    else:
        plx1.append(x[0])
        ply1.append(x[1])
    loss += -((1-Y[i])*np.log(1-op[0])+ (Y[i])*np.log(op[0]))
print(count)
print(loss)

plt.scatter(plx, ply, c='b')
plt.scatter(plx1, ply1, c='r')
```
