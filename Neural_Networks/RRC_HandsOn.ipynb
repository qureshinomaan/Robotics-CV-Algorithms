```python
from __future__ import print_function 
import torch 
%matplotlib inline 
```

```python
x = torch.empty(5, 3)
print(x)
```

```python
# gaussian, xavier 
x = torch.rand(5, 3)   # Uniform distribution between 0 and 1
print(x)
```

```python
x = torch.zeros(5, dtype=torch.long)
print(x)
```

```python
x = torch.tensor([[5., 3], [2, 1]])
print(x)
```

```python
x = torch.rand(5, 3)
y = torch.rand(5, 3)
torch.add(x, y)
y.add(x)
```

```python
# torch.view, torch.resize, torch.reshape
if torch.cuda.is_available() == True:
    device = torch.device("cuda")
    y = torch.ones_like(x, device=device)
```
